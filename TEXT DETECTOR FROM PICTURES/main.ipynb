```python
import cv2
import easyocr as ocr

import matplotlib.pyplot as plt



```

```python
image_path='test1.png'
img=cv2.imread(image_path)

#instantiate OCR library
detector=ocr.Reader(['en'], gpu=False)


```

```python
#retreive text from the image
txt=detector.readtext(img)

for bbox, text, temperature in txt:
    if temperature>=0.3:
        print(text)
        cv2.rectangle(img, bbox[0],bbox[2],(0,0, 255), 4)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()


###NB for streamlit, the user should be able to choose the temperature...
```
